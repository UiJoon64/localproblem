```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

df = pd.read_csv("부산광역시_노인교실 현황_20230930.csv", encoding='cp949')
data = gpd.read_file("HangJeongDong_ver20230701.geojson")

# 노인교실의 행정동 찾기
df["행정동명"] = ""  # 새로운 컬럼 "행정동명" 추가

for index, center in df.iterrows():
    point = Point(float(center["경도"]), float(center["위도"]))  # 숫자로 변환하여 사용
    for _, row in data.iterrows():
        if row["geometry"].contains(point):
            df.at[index, "행정동명"] = row["adm_nm"]  # 행정동명(adm_nm)을 새로운 컬럼에 할당
            break

# 결과를 확인하기 위해 데이터프레임 출력
print(df.head())
```

```python
df.to_csv("updated_senior_school.csv",encoding='cp949')
```
